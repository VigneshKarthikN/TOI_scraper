```python
from requests import get
from bs4 import BeautifulSoup 
import csv
import pandas as pd
```

```python
with open('toi_AA_INDIA_data.csv', 'w') as file:  # To delete the file content initially   
    writer = csv.writer(file)
    row = ['Heading', 'Publisher', 'Date', 'Body']
    writer.writerow(row)
```

```python
def clean(text):
    text = text.replace('\n', '')
    text = text.replace('\t', '')
    return text
```

```python
rows = []
Url = 'https://timesofindia.indiatimes.com/topic/Alcoholics-Anonymous-India/'
for pages in range(1, 10):
    url = Url + str(pages) # retrived all 9 pages
    response = get(url)
    response = BeautifulSoup(response.text, 'html.parser')
    containers = response.find_all('div', class_="tab_content")
    lists = containers[0].find_all('li',class_='article')

    for new_urls in lists:
        new_url = "https://timesofindia.indiatimes.com" + new_urls.find('a')['href']
        new_response = get(new_url)
        new_response = BeautifulSoup(new_response.text, 'html.parser')
        try:
            heading = new_response.find('h1', class_="heading1").text ######
            print(new_url)
            publisher = new_response.find_all('span', class_="time_cptn")[0].find_all('span')
            date = publisher[-1].text ######
            publisher = publisher[0].text ######
            body = clean(new_response.find('div', class_="Normal").text) ######
        except:
            pass
        rows.append(heading)
        rows.append(publisher)
        rows.append(date)
        rows.append(body)
        with open('toi_AA_INDIA_data.csv', 'a') as file:
            writer = csv.writer(file)
            writer.writerow(rows)
            rows = []
            body = ''
            file.close()       
```
